```python
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
heartDataset = pd.read_csv('Heart Attack.xls', delimiter=';')
```

```python
heartDataset.head()
```

```python
heartDataset['class'] = pd.factorize(heartDataset['class'])[0]
```

```python
heartDataset.head()
```

```python
print(heartDataset)
```

```python
#heartDataset = heartDataset.drop(labels=['id'], axis=1)   digunakan jika ada label Id pada dataset,
#guna nya untuk menghapus label Id karena tidak digunakan
```

```python
heartDataset = heartDataset.to_numpy()
```

```python
dataTraining = np.concatenate((heartDataset[0:40, :], heartDataset[50:90, :]), axis=0)
dataTesting = np.concatenate((heartDataset[40:50, :], heartDataset[90:100, :]), axis=0)
```

```python
inputTraining = dataTraining[:, 0:7]
inputTesting = dataTesting[:, 0:7]
labelTraining = dataTraining[:, 7]
labelTesting = dataTesting[:, 7]
print(labelTraining)
len(labelTraining)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(inputTraining, labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar :", prediksiBenar, "data")
print("Prediksi Salah :", prediksiSalah, "data")
print("Akurasi :", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
